```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline
```

<p><b>Задача 2.</b> Анализ рынка акций</p>

<p>1. Загрузите данные в один dataframe из всех файлов в папке /data/stock. Все файлы имеют одинаковую структуру, в том числе наименование столбцов. В качестве значений индекса dataframe'а необходимо указать значения столбца "Date". Название столбцов должные соответствовать названию акций (имя файла без .csv), а их значения - значениям цены закрытия (столбец "Close" в файлах .csv)</p>

```python
#Ваш код
import os

main_path = "./data/stock"

paths = os.listdir(main_path)

all_data = pd.DataFrame()
for i in range(0,len(paths)):
    curr_path = main_path + "/" + paths[i]
    curr_name = paths[i][0:-4] 
    
    # read current file data and add buffer column we want to get 
    file_data = pd.read_csv(curr_path, index_col=0)
    file_data[curr_name] = file_data["Close"]
    
    # add needed column to all data
    all_data = pd.concat([all_data,pd.DataFrame(file_data[curr_name])], axis=1)
    
all_data
```

<p>2. Рассчитайте корреляционную матрицу для всех акций</p>

```python
corr_matrix = all_data.corr()
corr_matrix
```

<p>3. Отобразите корреляционную матрицив в виде диаграммы. Примерный вид графика приведен ниже</p>
<img src="img/fig_corr_matrix.png" width="400px">

```python
fig = plt.figure(figsize=(15,15))
subplot = fig.add_subplot(111)
cax = subplot.matshow(corr_matrix, vmin=-1, vmax=1)
subplot.xaxis.set_major_locator(mpl.ticker.MultipleLocator(2))
subplot.yaxis.set_major_locator(mpl.ticker.MultipleLocator(2))
subplot.set_xticklabels(all_data.head(0))
subplot.set_yticklabels(all_data.head(0))
plt.show()
```

```python
# facebook
min_value = corr_matrix["FB"].drop("FB").idxmin()
max_value = corr_matrix["FB"].drop("FB").idxmax()
av_value = corr_matrix["FB"].drop("FB").abs().idxmin()

min_value, max_value, av_value
```

<p>5. Постройте диаграммы разброса (Ваша компания - Компания с min), (Ваша компания - Компания с max), (Ваша компания - Компания с none)</p>

```python
plt.title("max")
plt.scatter(corr_matrix["FB"], corr_matrix[max_value], s=25)
plt.show()

plt.title("min")
plt.scatter(corr_matrix["FB"], corr_matrix[min_value], s=25)
plt.show()

plt.title("none")
plt.scatter(corr_matrix["FB"], corr_matrix[av_value], s=25)
plt.show()
```

<p>6. Рассчитайте среднюю цену акций для каждого месяца (исходные данные взяты с интервалом в месяц)</p>

```python
all_data["Average"] = all_data.mean(axis=1)
all_data
```

<p>7. Постройте графики для акций из пункта 4 и средней из пункта 6. Примерный вид графика приведен ниже</p>
<img src="img/fig_tw_comp.png">

```python
plt.figure("222", figsize=(20,10))
plt.title("FB comparison")
plt.grid(True)

for i in range(0, 26):
    curr_row = all_data.iloc[i]
    
    plt.scatter(i, curr_row["FB"], c='blue')
    plt.scatter(i, curr_row[max_value], c='orange')
    plt.scatter(i, curr_row[min_value], c='green')
    plt.scatter(i, curr_row[av_value], c='red')
    plt.scatter(i, curr_row["Average"], c='grey')
    
    
    if i > 0:
        prev_row = all_data.iloc[i-1]
        plt.plot([i-1, i],[prev_row["FB"], curr_row["FB"]], marker='o', c='blue')
        plt.plot([i-1, i],[prev_row[max_value], curr_row[max_value]], marker='o', c='orange')
        plt.plot([i-1, i],[prev_row[min_value], curr_row[min_value]], marker='o', c='green')
        plt.plot([i-1, i],[prev_row[av_value], curr_row[av_value]], marker='o', c='red')
        plt.plot([i-1, i],[prev_row["Average"], curr_row["Average"]], marker='o', c='grey')


indexes = all_data.index.values
plt.xticks([0, 5, 10, 15, 20, 25],[indexes[0], indexes[5], indexes[10], indexes[15], indexes[20], indexes[25]])
plt.legend(['FB',max_value + " - max",min_value + " - min",av_value + " - null", "Average"])

plt.show()
```
