#### Задание 4 (2 балла)

Можете ли вы придумать правило для классификации вина на хорошее и плохое по рисунку выше? Пусть это будет нашей первой моделью)

Напишите функцию `theta = brute_clf_train(df)` которая бы перебирала пороговое значение $\theta$ по признаку `alcohol`, делало предсказание и выводило наилучшее $\theta$.<br/>
Т.е. если `alcohol` $\geq \theta$, то `prediction = 1` и `0` - иначе.


Оптимальность $\theta$ будем определять по доле совпадения значений `prediction` и `quality_cat` которое оно дает.

```python
# Заметим, что бОльшая часть "плохого" вина
# имеет около 10% алкоголя в составе. Качественное
# же вино распределено более равномерно. Скорее всего,
# некачественное вино имеет часто менее 11% алкоголя 
# или около того.


# Функция для вычисления наилучшего параметра "Тэта"
def brute_clf_train(df):
    row_count = wine_data.shape[0] - 1
    best_theta = -1.0 # Что находим
    best_count = -1   # Лучшее значение совпадения
    
    for theta in range(80, 141): # Проверим от 8 до 14 включительно
        curr_theta = float(theta) / 10.0
        curr_count = 0
        
        for i in range(0,row_count):
            curr_obj = df.loc[i]
            curr_count += 1 if ((curr_obj['alcohol'] >= curr_theta) == (curr_obj['quality_cat'] == 1)) else 0
        
        if curr_count >= best_count:
            best_count = curr_count
            best_theta = curr_theta

    return best_theta


best_theta = brute_clf_train(wine_data)  
best_theta
```

Проверим, как обобщается наша модель на другие данные.

```python
# Загрузим новые данные
white_wine_data = pd.read_csv('C:\Users\Danil\Desktop\IAD\sem\HW0\winequality-white.csv', sep=';')

# Добавим столбец качества, как мы уже делали
def get_category(quality):
    return (1 if quality > 5 else 0)

white_wine_data["quality_cat"] = map(get_category, white_wine_data["quality"])


# Функция подсчета процента верных предсказаний
def brute_clf_predict(df, theta):
    count = df.shape[0] - 1
    summ = 0
    for i in range(0,count):
        summ += 1 if ((df.loc[i]['alcohol'] >= theta) == (df.loc[i]['quality_cat'] == 1)) else 0
    return (float(summ) / count) 

# Вычислим процент верных предсказаний
brute_clf_predict(white_wine_data, best_theta)
```
